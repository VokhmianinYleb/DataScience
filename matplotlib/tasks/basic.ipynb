```python
# Task 1

import matplotlib.pyplot as plt

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')

plt.plot(range(1, 50), [value * 3 for value in range(1, 50)])

plt.show()
```

```python
# Task 5

import matplotlib.pyplot as plt

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two or more lines on same plot with suitable legends ')

plt.plot([10, 20, 30], [20, 40, 10], label = "line 1")
plt.plot([10, 20, 30], [40, 10, 30], label = "line 2")

plt.legend()

plt.show()
```

```python
# Task 6

import matplotlib.pyplot as plt

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two or more lines on same plot with suitable legends ')

plt.plot([10, 20, 30], [20, 40, 10], color = "blue", label = "line 1")
plt.plot([10, 20, 30], [40, 10, 30], color = "red", label = "line 2")

plt.legend()

plt.show()
```

```python
# Task 7

import matplotlib.pyplot as plt

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two or more lines on same plot with suitable legends ')

plt.plot([10, 20, 30], [20, 40, 10], color = "blue", label = "line 1", linestyle = 'dotted')
plt.plot([10, 20, 30], [40, 10, 30], color = "red", label = "line 2", linestyle = 'dashed')

plt.legend()

plt.show()
```
