```python
with open('Task.pkl', 'rb') as file:
    data = pickle.load(file)

mesh_list=list(data.loc[0, 'mesh'])
```

```python
len(mesh_list)
main_dui_list=[]
main_mesh_list=[]

for i in data.index:
    list_dict=list(data.loc[i,'mesh'])
    all_dui_list=[]
    all_mesh_list=[]
    for item in list_dict:
        all_dui_list.append(item['descriptor_ui'])
        all_mesh_list.append(item['descriptor_name'])
    main_dui_list.append(all_dui_list)
    main_mesh_list.append(all_mesh_list)
```

```python
len(main_mesh_list)
```

```python
data["ALL_DUI"] = main_dui_list
data["ALL_MESH"] = main_mesh_list
```
